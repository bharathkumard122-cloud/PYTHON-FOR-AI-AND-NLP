```python
import pandas as pd
import requests
import os

# Define the URL and local file path for nba.csv
url = "https://raw.githubusercontent.com/ashishpatel26/Pandas-Numpy-Project-Datasets/main/nba.csv"
file_path = "/content/nba.csv"

# Download the nba.csv file using requests for better error handling
try:
    response = requests.get(url, stream=True)
    response.raise_for_status() # Raise an exception for HTTP errors (4xx or 5xx)

    with open(file_path, 'wb') as f:
        for chunk in response.iter_content(chunk_size=8192):
            f.write(chunk)
    print(f"File downloaded successfully to {file_path}")

except requests.exceptions.RequestException as e:
    print(f"Error downloading file: {e}")
    raise # Re-raise the exception to stop execution if download fails

# Now read the file with pandas
data = pd.read_csv(file_path, index_col ="Name")

first = data.loc["Avery Bradley"]
second = data.loc["R.J. Hunter"]

print(first, "\n\n\n", second)
```
