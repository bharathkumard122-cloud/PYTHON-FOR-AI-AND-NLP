```python
import nltk
from nltk import ngrams
from collections import Counter
```

```python
nltk.download('punkt')
```

```python
text = """Natural Language Processing enables machines to understand human language.
It is widely used in chatbots, translation, and sentiment analysis."""
```

```python
tokens = nltk.word_tokenize(text.lower())
print("Tokens:\n", tokens)
```

```python
unigrams = list(ngrams(tokens, 1))
print("\nUnigrams:\n", unigrams)
```

```python
bigrams = list(ngrams(tokens, 2))
print("\nBigrams:\n", bigrams)
```

```python
trigrams = list(ngrams(tokens, 3))
print("\nTrigrams:\n", trigrams)
```

```python
uni_freq = Counter(unigrams)
bi_freq = Counter(bigrams)
tri_freq = Counter(trigrams)
```

```python
print("\nMost Common Unigrams:")
for gram, freq in uni_freq.most_common(5):
    print(gram, ":", freq)

print("\nMost Common Bigrams:")
for gram, freq in bi_freq.most_common(5):
    print(gram, ":", freq)

print("\nMost Common Trigrams:")
for gram, freq in tri_freq.most_common(5):
    print(gram, ":", freq)
```

```python
with open("ngram_results_ii.txt", "w") as file:
    file.write("Unigrams:\n")
    for gram, freq in uni_freq.items():
        file.write(f"{gram} : {freq}\n")

    file.write("\nBigrams:\n")
    for gram, freq in bi_freq.items():
        file.write(f"{gram} : {freq}\n")

    file.write("\nTrigrams:\n")
    for gram, freq in tri_freq.items():
        file.write(f"{gram} : {freq}\n")

print("Results saved in ngram_results_ii.txt")
```
